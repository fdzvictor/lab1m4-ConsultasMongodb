### 2. Todas las empresas que tienen más de 5000 empleados. Limita la búsqueda a 20 empresas y ordénalas por **número de empleados**.

```python

query2 = empresas.find({'number_of_employees': { "$eq": 5000}}).distinct("name").sort('number_of_employees').limit(20)
```

```python
query2 = empresas.find(
    {'number_of_employees': { "$eq": 5000 }},  
    {'name': 1, 'number_of_employees': 1}      
).sort('number_of_employees', 1).limit(20)     

# Extraer nombres únicos usando un conjunto
nombres_unicos = list({empresa['name'] for empresa in query2})
nombres_unicos
```
